```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
print(y.shape)
y = le.fit_transform(y)
y
```

```python
#creates models
def create_model(no_HL, HL_func, O_func):
    units = [8, 16, 32, 64, 128, 256]
    model = tf.keras.models.Sequential()
    model.add( tf.keras.layers.Flatten() )

    for i in range(no_HL):

        model.add( tf.keras.layers.Dense( units=units[i] , activation=HL_func ))

    model.add(tf.keras.layers.Dense(units=1, activation = O_func))

    return model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#creates a group of models
def create_model_set(no_HL):
    model_l = []
    split_set = []

    #different hidden layer functions to test with
    HL_func = ['relu', 'leaky', 'relu', 'leaky', 'relu', 'leaky', 'relu', 'leaky', 'relu']

    #different output layer functions to test with
    O_func = ['sigmoid', 'tanh', 'softmax', 'sigmoid', 'tanh', 'softmax', 'sigmoid', 'tanh', 'softmax',]

    #different optimizers to test with
    optim = ['adam', 'adam', 'adam', 'RMSprop', 'RMSprop', 'RMSprop', 'Adagrad', 'Adagrad', 'Adagrad']

    #different random states to test with 
    rand_states = [0, 10, 15, 20, 25, 30, 35, 40, 45]

    #range(9) cus i made it 9 models per table
    for i in range(9):
        if HL_func[i] == 'leaky':
            model = create_model(no_HL, tf.keras.layers.LeakyReLU(alpha=0.3), O_func[i])
            model.compile(optimizer = optim[i], loss = 'binary_crossentropy', metrics = ['accuracy'])
            model_l.append(model)
        else:
            model = create_model(no_HL, HL_func[i], O_func[i])
            model.compile(optimizer = optim[i], loss = 'binary_crossentropy', metrics = ['accuracy'])
            model_l.append(model)

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = rand_states[i])
        split_set.append((X_train, X_test, y_train, y_test))
        
    return model_l, split_set
    
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
#trains models
def fit_models(models, split_sets, epoch, batch_size, table_no):
    accuracies = []
    for idx, (model, split_set) in enumerate(zip(models, split_sets)):
        sc = StandardScaler()
        X_train, X_test, y_train, y_test = split_set
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)
        model.fit(X_train, y_train, batch_size = batch_size, epochs = epoch)

        y_pred = model.predict(X_test)
        y_pred = (y_pred > 0.5) 
        cm = confusion_matrix(y_test, y_pred)
        score = accuracy_score(y_test, y_pred)
        accuracies.append(score)

        print(f'model {idx}, of table {table_no} is done')
    
    return accuracies
        
```

```python
epochs = [100, 200, 300, 400, 500 ,600]
batch_sizes = [16, 32, 64, 128, 256, 256]
accuracies_list = []

#six cus i had 6 tables last time
for i in range(6):  
    models, split_sets = create_model_set(i+1)
    accuracies = fit_models(models, split_sets, epochs[i], batch_sizes[i], i)
    accuracies_list.append(accuracies)
    with open('accuracies3.txt', 'a') as f:
        f.write('\n'+str(accuracies))
```
